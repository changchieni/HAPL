```python
def h():
  h=(int(input('聖誕樹層數:')))
  a=''
  for i in range(h):
    a+=((' ' * (h-i)) + ('*' *((2*i)+1))+'\n')

  h2=5
  b=''
  for i in range(h2):
    b+=(('  '*(h2-1))+'*'*3+'\n')
  return a+b
print(h())
```
