```python
a=float(input('請輸入一個年份:'))
count=a%4
count1=a%100
count2=a%400
if a<1582 or a==1582 :
  if count==0 :
    print('此年為閏年')
  else :
    print('此年為平年')
else:
  if count==0 :
    if count1==0 and count2!=0 :
      print('此年為平年')
    elif count2==0 :
      print('此年為閏年')
  else :
    print('此年為平年')
```
