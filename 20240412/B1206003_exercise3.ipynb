```python
h=float(input("請輸入身高(cm):"))
w=float(input("請輸入體重(kg):"))
h/=100
h*=h
w/=h
BMI=w
print("BMI=",BMI)
if BMI<18.5 :
  print('您的體位為:過輕')
elif BMI<24 :
  print('您的體位為:適中')
elif BMI<27 :
  print('您的體位為:過重')
else :
  print('您的體位為:肥胖')
```

```python
a=float(input('請輸入第一個數字'))
b=str(input('請輸入運算符號'))
c=float(input('請輸入第二個數字'))
if b==('+') :
  print(a+c)
elif b==('-') :
  print(a-c)
elif b==('*') :
  print(a*c)
else :
  print(a/c)

```
